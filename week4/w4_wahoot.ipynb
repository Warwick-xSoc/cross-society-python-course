## Reading and parsing questions

The Wahoot code is roughly separated into two directories: `wahoot_web`, which takes care of the our interface or web application (e.g. the menu, buttons, leaderboard pages, submitting answers, ...), and `wahoot`, which includes the core of the logic behind Wahoot (for example, calculating the score, keeping track of players, ranking players, ...).

For the purposes of our course we will focus on `wahoot`, however, if you're curious you can also explore `wahoot_web`. 

Within `wahoot` we have separated loading our data (questions or game history, for example) from the business logic.
This makes our code more flexible (as we will see in Week 5), and improves maintanability and readability.
In particular, code residing in `repositories.py` is concerned *only* with loading data from somewhere (e.g. files, memory, database) which leads us into our first task.


> **Task 1**: Implement the `__init____` method within the `QuestionBank` class in `repositories.py` which should take the path of a CSV file and return a list of questions.

In Wahoot the questions are stored in a CSV file as follows:

```csv
difficulty,text,correct,wrong_1,wrong_2,wrong_3
0,How many societies are involved in this course?,3,1,2,4
0,Is Wahoot a completely original concept?,Yes,No,Nope,Not at all
1,How many weeks of content are there on the course?,6,4,5,7
2,Which society is not involved in this course?,Quidditch,AI,Coding,Computing
2,Which of these data structures is not taught on this course?,Sets,Lists,Tuples,Dictionaries
```

We store the difficulty, the actual question or text, the correct answer, and the wrong answers.

As you have probably noticed, this method is in a class and it contains a parameter named `self`. These are object-oriented concepts which will be introduced in Week 5. For now, you can think of classes as a way of organizing your code into meaningful sections, improving its maintanability and reusability. You can ignore `self` and the fact that the method is in the `QuestionBank` class for now, and implement the method as usual.
